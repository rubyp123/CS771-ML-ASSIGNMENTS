```python
#part 3 
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']

with open('mnist_small.pkl', 'rb') as file:
    data = pickle.load(file)

X = data['X']  
Y = data['Y']  

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 5))
for i in range(10):  
    idx = np.where(np.array(Y) == i)[0]
    x = [X_pca[idx][0] for idx in idx]
    y = [X_pca[idx][1] for idx in idx]
    plt.scatter(x, y, marker='o', s=20, color=colors[i], label=f'Digit {i}', alpha=0.7)

plt.title('PCA Projection')
plt.legend()
plt.show()

tsne = TSNE(n_components=2, perplexity=30, n_iter=400)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10, 5))
for i in range(10): 
    idx = np.where(np.array(Y) == i)[0]
    x = [X_tsne[idx][0] for idx in idx]
    y = [X_tsne[idx][1] for idx in idx]
    plt.scatter(x, y, marker='o', s=20, color=colors[i], label=f'Digit {i}', alpha=0.7)

plt.title('t-SNE Projection')
plt.legend()
plt.show()
```
